```python
df.Make.value_counts().nlargest(100).plot(kind='bar', figsize=(12,5))
plt.title('Top Car Makers')
plt.xlabel('Car Makers')
plt.ylabel('Number of Cars')
plt.show()
```

```python
# concatinating two columns in df
df['Make_Model'] = df['Make'] + ' ' + df['Model']
```

```python
def get_price_range(msrp):
    if msrp < 20000:
        return 'Low'
    elif 20000 <= msrp < 40000:
        return 'Medium'
    else:
        return 'High'
```

```python
df['Price_Range'] = df['MSRP'].apply(get_price_range)
df.head()
```
